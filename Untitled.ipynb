```python
import os
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
from pydub import AudioSegment
```

```python
audio_dir = "data/raw_data"
```

```python
audio_files = [f for f in os.listdir(audio_dir) if f.endswith(".mp3")]
print("Audio files found: \n")
for f in audio_files:
    print(f)
```

```python
a_data = {}

for audio in audio_files:
    path = os.path.join(audio_dir, audio)
    y, sr = librosa.load(path, sr = 16000)
    a_data[audio] = (y, sr)
    print(audio, "loaded | length:", round(len(y)/sr , 2), "seconds")
```

```python
file  = audio_files[0]
y, sr = a_data[file]

plt.figure(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr)
plt.title(f"Waveform: {file}")
plt.xlabel("Time")
plt.ylabel("amp")

plt.show()
```

```python
trim_times = {
    "038010_EIT-2A.mp3":160,
    "038015_EIT-1A.mp3":150,
    "038012_EIT-2A.mp3":739,
    "038011_EIT-1A.mp3":147
}
```

```python
after_trim = {}

for file, (y, sr) in a_data.items():
    start = trim_times[file]
    trimd = y[start*sr:]
    after_trim[file] = (trimd, sr)
    print(file, "trimmed | new length: ", round(len(trimd)/sr, 2))
```

```python
from pydub.silence import detect_nonsilent
import io
```

```python
def numpy_seg(y, sr):
    bytes = (y*32767).astype(np.int16).tobytes()
    seg = AudioSegment(data = bytes, sample_width = 2, frame_rate = sr, channels = 1)
    return seg
```

```python
all_segs = {}
for n, (y, sr) in after_trim.items():
    print("#n Processing: ", n)

    seg = numpy_seg(y, sr)

    segments = detect_nonsilent(seg, min_silence_len = 4000, silence_thresh = seg.dBFS - 10)

    print("Segements Detected: ", len(segments))
    all_segs[n] = segments
```

```python
for n, (y, sr) in after_trim.items():
    segments = all_segs[n]
    plt.figure(figsize = (14, 5))
    librosa.display.waveshow(y, sr = sr)

    for start, end in segments:
        ss = start / 1000
        es = end / 1000

        plt.axvspan(ss, es, color = "blue", alpha = 0.3)

    plt.title(f"Detected Speech segments: {n}")
    plt.show()
```
